## Basic .wav file manipulations

```python
import IPython.display as ipd
ipd.Audio('../data/Yamaha.wav')
```

```python
from scipy.io import wavfile

sr, data = wavfile.read('../data/Yamaha.wav')
```

```python
data
```

```python
import librosa
import librosa.display as disp


data, sampling_rate = librosa.load('../data/Yamaha.wav')
```

```python
data
```

```python
sampling_rate
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(12,4))

disp.waveplot(data, sr=sampling_rate)
```

```python
S = np.abs(librosa.stft(data))
disp.specshow(librosa.power_to_db(S**2, ref=np.max), sr=sr, x_axis='time', y_axis='linear')
plt.colorbar()
```

```python
from scipy.fftpack import fft

fft_raw = fft(data)
print("length of audio: {}, length of fft: {}".format(len(data), len(fft_raw)))
plt.plot(fft_raw)
plt.title("Raw FFT")
plt.show()
```

```python
t = np.linspace(0, 4, int(4*sr), endpoint=False)
x = 0.5*np.sin(2*np.pi*440*t)
librosa.output.write_wav('../data/tone_440.wav', x, sr)
```

```python
ipd.Audio('../data/tone_440.wav')
```
